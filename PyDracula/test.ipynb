```python
import sys
from PySide6.QtWidgets import (
    QApplication, QWidget, QHBoxLayout, QVBoxLayout,
    QListWidget, QListWidgetItem, QLabel, QFrame
)
from PySide6.QtGui import QPixmap, QColor
from PySide6.QtCore import Qt, QSize

# 각 리스트 항목에 대한 데이터
ITEMS_DATA = [
    {
        "name": "항목 1 (서울)",
        "image_path": None, # "path/to/your/image1.png"
        "latitude": 37.5665,
        "longitude": 126.9780
    },
    {
        "name": "항목 2 (부산)",
        "image_path": None,
        "latitude": 35.1796,
        "longitude": 129.0756
    },
    {
        "name": "항목 3 (대구)",
        "image_path": None,
        "latitude": 35.8714,
        "longitude": 128.6014
    },
    {
        "name": "항목 4 (인천)",
        "image_path": None,
        "latitude": 37.4563,
        "longitude": 126.7052
    },
]

# 1. 왼쪽 리스트의 각 항목을 구성할 커스텀 위젯 클래스
class CustomListItem(QWidget):
    def __init__(self, item_data, parent=None):
        super().__init__(parent)
        self.item_data = item_data
        self.init_ui()

    def init_ui(self):
        layout = QHBoxLayout(self)
        layout.setContentsMargins(5, 5, 5, 5) # 위젯 내부 여백

        # 이미지 라벨
        image_label = QLabel()
        image_label.setFixedSize(60, 60)
        
        image_path = self.item_data.get("image_path")
        if image_path:
            pixmap = QPixmap(image_path)
        else:
            # 이미지가 없을 경우, 이름에 따라 색상 생성
            pixmap = QPixmap(60, 60)
            color = QColor((hash(self.item_data['name']) & 0x00FFFFFF) | 0x00A0A0A0)
            pixmap.fill(color)
            
        image_label.setPixmap(pixmap.scaled(
            image_label.size(),
            Qt.AspectRatioMode.KeepAspectRatio,
            Qt.TransformationMode.SmoothTransformation
        ))
        
        # 텍스트 정보 (이름, 위도, 경도)
        text_info_layout = QVBoxLayout()
        name_label = QLabel(self.item_data["name"])
        name_label.setStyleSheet("font-weight: bold;")
        
        coords_label = QLabel(f"위도: {self.item_data['latitude']:.4f}, 경도: {self.item_data['longitude']:.4f}")
        coords_label.setStyleSheet("font-size: 9pt; color: gray;")

        text_info_layout.addWidget(name_label)
        text_info_layout.addWidget(coords_label)
        text_info_layout.addStretch()

        layout.addWidget(image_label)
        layout.addLayout(text_info_layout)


class MainApp(QWidget):
    def __init__(self):
        super().__init__()
        self.init_ui()

    def init_ui(self):
        self.setWindowTitle('Qt 커스텀 리스트 아이템 예제')
        self.setGeometry(300, 300, 900, 450)

        main_layout = QHBoxLayout(self)

        # 2. 왼쪽: 리스트 위젯 생성
        self.list_widget = QListWidget()
        self.list_widget.setMinimumWidth(350) # 왼쪽 리스트의 최소 너비 설정

        # 데이터 채우기
        for item_data in ITEMS_DATA:
            # 빈 QListWidgetItem 생성
            list_item = QListWidgetItem(self.list_widget)
            
            # 우리가 만든 커스텀 위젯 생성
            custom_widget = CustomListItem(item_data)
            
            # 아이템의 크기를 커스텀 위젯의 크기에 맞춤
            list_item.setSizeHint(custom_widget.sizeHint())
            
            # UserRole에 원본 데이터를 저장 (선택 시 상세 뷰 업데이트를 위해 필요)
            list_item.setData(Qt.ItemDataRole.UserRole, item_data)
            
            # 리스트 위젯에 아이템을 추가하고, 해당 아이템 위치에 커스텀 위젯을 설정
            self.list_widget.addItem(list_item)
            self.list_widget.setItemWidget(list_item, custom_widget)
        
        # 3. 오른쪽: 상세 정보 뷰 (이전 코드와 동일)
        details_container = QFrame()
        details_container.setFrameShape(QFrame.Shape.StyledPanel)
        details_layout = QVBoxLayout(details_container)

        self.image_label = QLabel("이미지 없음")
        self.image_label.setAlignment(Qt.AlignmentFlag.AlignCenter)
        self.image_label.setFixedSize(300, 200)
        self.image_label.setStyleSheet("background-color: #f0f0f0; border: 1px solid #ccc;")

        self.details_title_label = QLabel("항목 박스 구성")
        self.details_title_label.setStyleSheet("font-size: 16px; font-weight: bold; margin-bottom: 10px;")
        
        self.latitude_label = QLabel("위도: --")
        self.longitude_label = QLabel("경도: --")
        
        details_layout.addWidget(self.details_title_label)
        details_layout.addWidget(self.image_label)
        details_layout.addWidget(self.latitude_label)
        details_layout.addWidget(self.longitude_label)
        details_layout.addStretch(1)

        # 4. 전체 레이아웃에 위젯 추가
        main_layout.addWidget(self.list_widget)
        main_layout.addWidget(details_container, 1)

        # 5. 시그널-슬롯 연결
        self.list_widget.currentItemChanged.connect(self.update_details)
        
        if self.list_widget.count() > 0:
            self.list_widget.setCurrentRow(0)

    def update_details(self, current_item, previous_item):
        if current_item is None:
            return

        item_data = current_item.data(Qt.ItemDataRole.UserRole)

        image_path = item_data.get("image_path")
        if image_path:
            pixmap = QPixmap(image_path)
            self.image_label.setPixmap(pixmap.scaled(
                self.image_label.size(), 
                Qt.AspectRatioMode.KeepAspectRatio, 
                Qt.TransformationMode.SmoothTransformation
            ))
        else:
            placeholder_pixmap = QPixmap(self.image_label.size())
            color = QColor((hash(item_data['name']) & 0x00FFFFFF) | 0x00A0A0A0)
            placeholder_pixmap.fill(color)
            self.image_label.setPixmap(placeholder_pixmap)

        self.latitude_label.setText(f"위도: {item_data['latitude']}")
        self.longitude_label.setText(f"경도: {item_data['longitude']}")


if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = MainApp()
    ex.show()
    sys.exit(app.exec())
```
